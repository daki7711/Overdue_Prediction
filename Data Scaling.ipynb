```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
# The data we will be using
data = pd.read_excel("자료A0_상장비상장(2008-2013).xlsx",sheet_name='2008-2013 (kis다운받은변수추가후)',encoding='euc-kr')
```

```python
# This is how the data looks like
# To give a few columns They are "Company Name, Listed Company, Year, KIS Credit Rate, KDB Credit Rate, ... ,Total capital operating margin
# Equity Capital Ratio, Annual Report ..."
data.head()
```

```python
# This part is where we distinguish Discrete Variables (X1) and Continuous Variablesv(X2)
X1=data[['연도','상장여부', '취약기업여부','유명감사인','영업현금흐름','감사의견']]
X2=data[['KIS신용등급','kdb신용등급수정','자산건전성수정','부채비율 ','매출액영업이익율','LN(자산규모)',
         '매출액증가율','자기자본순익','순이익증가율','총자본회전율','총자본순이익율','총자산증가율','자기자본증가율',
         '영업이익증가율','총자본영업이익율','자본금순이익율','매출액순이익율','자기자본비율','자기자본회전율']]
```

```python
# Our Dependent variable will be The presence or absence of debt overdue for each company
y=data[['연체여부']]
```

```python
X1_dum=pd.get_dummies(X1)
F=pd.concat([X1_dum, X2, y], axis=1)

# This part is where we deal with missing values. For this part I filled in the missing values with median values
# However, this is part that needs further study
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values='NaN', strategy='median',axis=0)
imputer = imputer.fit(F)
imputed_data = imputer.transform(F.values)
```

```python
dataframe = pd.DataFrame.from_records(imputed_data)
dataframe=dataframe.dropna(subset=[36])
```

```python
dataframe.columns = F.columns
dataframe.head()
```

```python
# This is the final data csv file we will use throughout the whole project
dataframe.to_csv('상장비상장(2008_2013).csv', sep=',', encoding='euc-kr')
```
